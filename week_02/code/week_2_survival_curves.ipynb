# Load libraries


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min, norm, geom, uniform, lognorm
```

```python
# define useful variables
timesteps = np.arange(0, 61)
timesteps
```

# Create a survival curve

```python
# Fixed lifetime survival curve

fixed_lifetime = 40  # this parameters controls....
curve_surv = np.ones_like(timesteps)
curve_surv[fixed_lifetime:] = 0
plt.plot(curve_surv)
plt.show()
```

```python
# Uniform distribution
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html#scipy.stats.uniform

curve_surv = uniform.sf(
    timesteps,
    loc=10,  # this parameters controls....
    scale=20,  # this parameters controls....
)
plt.plot(curve_surv)
plt.show()
```

```python
# Geometric distribution
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.geom.html#scipy.stats.geom

curve_surv = geom.sf(
    timesteps,
    p=0.05,  # this parameters controls....
    loc=0,  # this parameters controls....
)
plt.plot(curve_surv)
plt.show()
```

```python
# Normal distribution
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm

curve_surv = norm.sf(
    timesteps,
    loc=30,  # this parameters controls....
    scale=10,  # this parameters controls....
)
plt.plot(curve_surv)
plt.show()
```

```python
# Weibull distribution
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.weibull_min.html

curve_surv = weibull_min.sf(
    timesteps,
    c=2,  # this parameters controls....
    loc=0,  # this parameters controls....
    scale=30,  # this parameters controls....
)
plt.plot(curve_surv)
plt.show()
```

```python
# Lognormal distribution
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.lognorm.html#scipy.stats.lognorm

curve_surv = lognorm.sf(
    timesteps,
    s=0.5,  # this parameters controls....
    loc=0,  # this parameters controls....
    scale=10,  # this parameters controls....
)
plt.plot(curve_surv)
plt.show()
```
